```python
import pandas as pd

games = pd.read_csv("cleaned_csv/games_clean.csv")
teams = pd.read_csv("cleaned_csv/teams_clean.csv")
```

```python
home = games[["team_id_home", "team_name_home", "team_abbreviation_home"]].rename(
    columns={
        "team_id_home": "team_id",
        "team_name_home": "team_name",
        "team_abbreviation_home": "abbreviation"
    }
)

away = games[["team_id_away", "team_name_away", "team_abbreviation_away"]].rename(
    columns={
        "team_id_away": "team_id",
        "team_name_away": "team_name",
        "team_abbreviation_away": "abbreviation"
    }
)

teams_fixed = pd.concat([home, away], ignore_index=True)

teams_fixed = teams_fixed.dropna(subset=["team_id"])
teams_fixed["team_id"] = pd.to_numeric(teams_fixed["team_id"], errors="coerce").astype("Int64")
teams_fixed = teams_fixed.drop_duplicates(subset=["team_id"])

teams_fixed.to_csv("cleaned_csv/teams_clean_fixed.csv", index=False)

print(teams_fixed.shape)
print(teams_fixed.head())
```
